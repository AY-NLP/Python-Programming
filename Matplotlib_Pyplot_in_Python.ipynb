Matplotlib is a python library. it is an open source drawing library.

Types of plots
* Bar chart
* Scatter plot
* Pie chart
* Histogram   
etc.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([1, 100]) # from 1 to 100
y = np.array([1, 50]) # from 1 to 50

plt.plot(x, y)
plt.show()
```

```python
x = np.array(['apple', 'mango', 'orange', 'coconut', 'avocado'])
y = np.array([300, 700, 200, 400, 2500])

plt.plot(x, y)
# Create a title
plt.title("Fruits Distribution")
# Create labels
plt.xlabel("Fruits")
plt.ylabel("Price")

plt.show()
```

```python
# Add grid lines to the plot

x = np.array(['apple', 'mango', 'orange', 'coconut', 'avocado'])
y = np.array([300, 700, 200, 400, 2500])

plt.plot(x, y)
# Create a title
plt.title("Fruits Distribution")
# Create labels
plt.xlabel("Fruits")
plt.ylabel("Prices")
#add grid lines
plt.grid()

plt.show()
```

Subplots

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 30, 50])
y = np.array([100, 2000, 300, 4000, 5000])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

Scatter Plot

```python
a = [10, 20, 30, 40, 50]
b = [100, 200, 300, 400, 500]

plt.scatter(a, b)
plt.show()
```

```python
y = [35, 25, 25, 15]
x = ["Apples", "Bananas", "Cherries", "Dates"]
plt.scatter(x, y, color = 'red')
plt.scatter(x, y)
plt.show()
```

Barplots

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([30, 80, 100, 10])

plt.bar(x, y, color = "green")
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.xlabel('Labels')
plt.ylabel('Numbers')
plt.title('Barplot')

plt.bar(x, y, width = 0.1, color= 'hotpink')
plt.show()
```

Histogram plot

```python
x= [1, 2, 3, 4, 5, 6, 7, 8, 10, 1, 6, 5, 6, 7, 8, 9, 10]
plt.hist(x)
plt.show()
```

```python
fruits = ['apple', 'mango', 'orange', 'banana', 'coconut', 'avocado']
quantity = [10, 20, 30, 40, 50, 60]

plt.pie(quantity, labels= fruits, autopct='%1.1f%%')
plt.show()
```
